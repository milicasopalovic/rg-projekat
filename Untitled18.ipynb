```python
import pandas as pd
```

```python
df = pd.read_csv('dogs.csv')
```

```python
df
```

```python
feature_names = df.columns[1:]
```

```python
feature_names
```

```python
X = df[feature_names]
```

```python
X
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X = scaler.fit_transform(X)
```

```python
X
```

```python
X = pd.DataFrame(X, columns = feature_names)
```

```python
X
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=3, linkage = 'single', compute_distances=True)
```

```python
model.fit(X)
```

```python
model.labels_
```

```python
df[model.labels_ == 2]
```

```python
model.children_.shape
```

```python
model.distances_
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(X, model.labels_)
```

```python
from matplotlib import pyplot as plt
```

```python
linkages = ['single', 'average', 'complete']
k = 3
fig = plt.figure(figsize = (15, 5))
for idx, linkage in enumerate(linkages):
  model = AgglomerativeClustering(n_clusters=k, linkage=linkage)
  model.fit(X)

  fig.add_subplot(1, len(linkages), idx + 1)

  for cluster_label in range(k):
    cluster = df[model.labels_ == cluster_label]
    plt.scatter(cluster['height'], cluster['weight'])

  plt.title(f'l = {linkage}, k = {k}, sil = {silhouette_score(X, model.labels_)}')

plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
z = linkage(X, method='single')
```

```python
z
```

```python
X = X.set_index(df['breed'])
```

```python
X
```

```python
_ = dendrogram(z, labels = X.index, leaf_rotation=90)
```

```python
from sklearn.cluster import DBSCAN
```

```python
model = DBSCAN(eps = 0.1, min_samples = 2)
```

```python
model.fit(X)
```

```python
model.labels_
```

```python
eps_values = [0.1, 0.15, 0.2, 0.25, 0.3]
min_sample_values = range(2, 5)
fig = plt.figure(figsize = (30, 10))
idx = 1
for eps in eps_values:
  for min_sample in min_sample_values:
    model = DBSCAN(eps = eps, min_samples = min_sample)
    model.fit(X)

    fig.add_subplot(len(eps_values), len(min_sample_values), idx)
    idx += 1

    k = max(model.labels_) +1
    for cluster_label in range(k):
      cluster = df[model.labels_ == cluster_label]
      plt.scatter(cluster['height'], cluster['weight'])
      plt.title(f'eps = {eps}, min_sample = {min_sample}, sil = {silhouette_score(X, model.labels_)}')
plt.tight_layout()
```
